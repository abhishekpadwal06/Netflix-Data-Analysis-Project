```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Analysis 1A: What is the percentage ratio of Movies to TV Shows on Netflix?
# PIE DIAGRAM

data = pd.read_csv('Dataset/netflix_titles.csv')
mov_type = data['type'].value_counts()

labels=['Movies', 'TV shows']

plt.pie(mov_type, labels=labels, shadow=True, wedgeprops={'edgecolor':'white'}, autopct='%1.1f%%')

plt.title("Movies v/s TV Shows on Netflix")
plt.tight_layout()

plt.savefig('images/fig1.png')
plt.show()
```

```python
# Analysis 1B: What is the absolute count difference between Movies and TV Shows on Netflix?
# BAR GRAPH

x_indexes = np.arange(1,)

data = pd.read_csv('Dataset/netflix_titles.csv')
mov_type = data['type'].value_counts()

plt.xlabel("Content Type")
plt.ylabel("Count")

plt.bar(mov_type.index, mov_type.values, width=0.7)

plt.title('Number of Movies and TV Shows on Netflix')
plt.tight_layout()

plt.savefig('images/fig2.jpg')
plt.show()
```

```python
# Analysis 2: Which are the top 10 countries that have the most content on Netflix
# BAR GRAPH

data = pd.read_csv('Dataset/Country_cleaned.csv')
data = data[data['Countries'] != 'Unknown']
country_counts = data['Countries'].value_counts().head(10)
country_counts = country_counts.sort_values()

plt.barh(country_counts.index, country_counts.values)
plt.xlabel('Count')
plt.ylabel('Countries')

plt.title("Top 10 Countries with the Most content on Netflix")

plt.tight_layout()
plt.savefig('images/fig3.jpg')
plt.show()
```

```python
# Analysis 3: Which are the top 10 genres on Netflix? 
data = pd.read_csv('Dataset/genre_cleaned.csv')
genre = data['genre'].value_counts().head(10)
genre = genre.sort_values()

plt.xlabel("Number of Movies/TV shows under specific genre")
plt.ylabel("Genre")

plt.barh(genre.index, genre.values)

plt.title('Top 10 Genres on Netflix')
plt.tight_layout()

plt.savefig('images/fig4.jpg')
plt.show()
```

```python
# Analysis 4: What is the change in the content added every year? 
# LINE GRAPH

data = pd.read_csv('Dataset/Year_and_Rating_cleaned.csv')
year = data['Year Added'].value_counts().sort_index()

plt.xlabel("Years")
plt.ylabel("Count")

plt.plot(year.index, year.values, marker='o')

plt.grid(alpha=0.3)
plt.xticks(rotation=45)
plt.title('Netflix Content Added Per Year')
plt.tight_layout()

plt.savefig('images/fig5.jpg')
plt.show()
```

```python
# Analysis 5: What is the distribution of ratings on Netflix?
# BAR GRAPH

data = pd.read_csv('Dataset/Year_and_Rating_cleaned.csv')
ratings = data['rating'].value_counts().head(10)

plt.bar(ratings.index, ratings.values)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Ratings')
plt.ylabel('Number of Content')
plt.xticks(rotation=30)
plt.tight_layout()

plt.savefig('images/fig6.jpg')
plt.show()
```

```python
# Analysis 6: Which are the top 10 most famous Actors/Actresses on Netflix (Having highest number of movies/shows)
# BAR GRAPH

data = pd.read_csv('Dataset/Cast_cleaned.csv')
data = data[data['Cast'] != "Unknown"]

cast = data['Cast'].value_counts().head(10)
cast = cast.sort_values()

plt.barh(cast.index, cast.values)

plt.title('Most Famous Actors on Netflix')
plt.xlabel('Number of Movies/Shows')
plt.ylabel('Actors')

plt.tight_layout()

plt.savefig('images/fig7.jpg')
plt.show()
```

```python
# Analysis 7: Which are the top 10 most famous Directors on Netflix (Having highest number of movies/shows)
data = pd.read_csv('Dataset/Cast_cleaned.csv')
data = data[data['director'] != "Unknown"]

dir = data['director'].value_counts().head(10)
dir = dir.sort_values()

plt.barh(dir.index, dir.values)

plt.title('Most Famous Directors on Netflix')
plt.xlabel('Number of Movies/Shows')
plt.ylabel('Directors')

plt.tight_layout()

plt.savefig('images/fig8.jpg')
plt.show()
```
